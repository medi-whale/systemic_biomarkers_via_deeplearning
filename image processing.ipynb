```python
import pandas as pd
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
from PIL import Image
import cv2
```

```python
img = Image.open("...jpg")
img
```

```python
def contrast_enhace(img) :
    scale = 300
    img = np.asarray(img)
    
    blur_img = cv2.GaussianBlur(img,(0 ,0) , scale/30)
    merge_img = cv2.addWeighted(img , 4 , blur_img , -4 , 128)
    b = np.zeros(img.shape)
    cv2.circle(b , (int(scale/2) ,int(scale/2)) , int(scale) , (1,1,1), -1 , 8 , 0 )
    merge_img = merge_img * b + 128 * (1 - b)
    merge_img = merge_img.astype('uint8')
    
    merge_img = cv2.cvtColor(merge_img,cv2.COLOR_BGR2RGB)

    return merge_img
```

```python
img_contrast = contrast_enhace(img)
```

```python
plt.figure()
plt.imshow(cv2.cvtColor(img_contrast,cv2.COLOR_BGR2RGB))
```
